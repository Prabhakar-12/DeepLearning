```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return x**2

# x values
x = np.arange(0, 3, 0.1)

# Plot the function
plt.plot(x, f(x), label='f(x) = x^2')

# Plot the tangent line
plt.plot(x, 2*x - 1, label='Tangent line at x=1', linestyle='--')

# Add labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('Function and Tangent Line')
plt.legend()

# Show plot
plt.show()
```
